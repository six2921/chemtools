## PAINS

```python
from rdkit.Chem import FilterCatalog
from rdkit.Chem.FilterCatalog import *
```

```python
params = FilterCatalogParams()
params.AddCatalog(FilterCatalogParams.FilterCatalogs.PAINS) # 마지막에 PAINS를 붙이면 PAINS 전체 탐색
```

```python
catalog = FilterCatalog(params)
mol = Chem.MolFromSmiles('OC1=CC=CC=C1O')
entry = catalog.GetFirstMatch(mol)
print(entry.GetDescription())

mol
```

```python
params = FilterCatalogParams()
params.AddCatalog(FilterCatalogParams.FilterCatalogs.PAINS_A) # 마지막에 PAINS_A를 붙이면 PAINS_A 전체 탐색
```

```python
mol = Chem.MolFromSmiles('O=C(Cn1cnc2c1c(=O)n(C)c(=O)n2C)N/N=C/c1c(O)ccc2c1cccc2')
entry = catalog.GetFirstMatch(mol)
print (entry.GetDescription())

mol
```

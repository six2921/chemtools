# Check PDB Ligand Info

```python
from check_pdb import pdb2ligands
```

```python
from check_pdb import uniprot2pdbligands
pdb_ligands = uniprot2pdbligands(uniprot_id, count_cutoff=100)
```

```python
pdb_ligands = pdb_ligands.loc[pdb_ligands['experimental_method'] == 'X-ray diffraction']
pdb_ligands = pdb_ligands.loc[pdb_ligands['numligs'] > 0]
pdb_ligands
```

```python
# pdb_ligands.to_csv('abc.csv', index=False)
```

```python
import rdkit.Chem.Draw as Draw
import rdkit.Chem as Chem

mols = [Chem.MolFromSmiles(x) for x in pdb_ligands['het0']]
Draw.MolsToGridImage(mols, legends=list(pdb_ligands['pdb_id']), useSVG=True)
```
